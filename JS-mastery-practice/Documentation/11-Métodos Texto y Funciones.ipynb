
# Métodos de Transformación de Textos en JavaScript

## 1. `indexOf()`
- **Descripción**: Devuelve la posición de la primera aparición de una subcadena en la cadena.
- **Uso**: `texto.indexOf("subcadena")`
- **Ejemplo**: 
  
          var busqueda = texto1.indexOf("curso"); // Devuelve 23
  
## 2. lastIndexOf()

Descripción: Devuelve la posición de la última aparición de una subcadena en la cadena.
Uso: texto.lastIndexOf("subcadena")
Ejemplo:

            var busqueda2 = texto1.lastIndexOf("de"); // Devuelve 32

            
## 3. search()
- Descripción: Busca una expresión regular en la cadena y devuelve la posición de la primera coincidencia.
- Uso: texto.search("palabra")
- Ejemplo:

            var busqueda3 = texto2.search("curso"); // Devuelve -1 (no encontrado)


## 4. match()
- Descripción: Busca coincidencias en la cadena según una expresión regular y devuelve detalles.
- Uso: texto.match(/expresiónRegular/g)
- Ejemplo:


            var busqueda4 = texto1.match("de"); // Detalles de la coincidencia


## 5. substring()
- Descripción: Devuelve una parte de la cadena entre dos índices.
- Uso: texto.substring(inicio, fin)
- Ejemplo:

              var busqueda6 = texto1.substring(14, 19); // Devuelve "curso"


## 6. charAt()
- Descripción: Devuelve el carácter en una posición específica de la cadena.
- Uso: texto.charAt(indice)
- Ejemplo:
  
            var busqueda7 = texto1.charAt(44); // Devuelve "R"
## 7. startsWith()
- Descripción: Comprueba si la cadena comienza con una subcadena específica, devolviendo true o false.
- Uso: texto.startsWith("subcadena")
- Ejemplo:

            var busqueda8 = texto1.startsWith("Victor"); // Devuelve false


## 8. includes()
- Descripción: Verifica si la cadena contiene una subcadena específica, devolviendo true o false.
- Uso: texto.includes("subcadena")
- Ejemplo:

  
            var busqueda8 = texto1.includes("Bienvenido"); // Devuelve true

## 9. replace()
- Descripción: Reemplaza la primera coincidencia de una subcadena por otra.
- Uso: texto.replace("antigua", "nueva")
- Ejemplo:

            var reemplazo = texto1.replace("Victor", "Ornella"); // Reemplaza "Victor" por "Ornella"


## 10. slice()
- Descripción: Extrae una sección de la cadena desde un índice específico.
- Uso: texto.slice(inicio, fin)
            - Ejemplo:

            var separación = texto1.slice(14); // Devuelve desde el índice 14 hasta el final
  
## 11. split()
-Descripción: Divide una cadena en un array de subcadenas usando un separador.
-Uso: texto.split("separador")
-Ejemplo:

            var splitieando = texto1.split(" "); // Devuelve un array de palabras
## 12. trim()

- Descripción: Elimina los espacios en blanco al inicio y al final de la cadena.
- Uso: texto.trim()
- Ejemplo:
            
            var borrarEspacios = texto3.trim(); // Elimina espacios sobrantes
