# Operaciones con Arrays


## Añadir y Eliminar Elementos

### Método .push()
- Descripción: Agrega un nuevo elemento al final del array.
- Uso:
        
        peliculas.push(elemento);

### Método .pop()

- Descripción: Elimina el último elemento del array.
- Uso:

        peliculas.pop();

  
## Búsqueda de Elementos

### Método indexOf()

- Descripción: Busca un elemento en el array y devuelve su índice. Si no se encuentra, devuelve -1.
- Uso:
        
        var indice = peliculas.indexOf('Batman');
        if (indice > -1) {
            peliculas.splice(indice, 1); // Elimina 'Batman' del array
        }

  
## Conversión entre String y Array

### Método .join()

- Descripción: Convierte todos los elementos de un array en una cadena, separados por comas.
- Uso:
        
        var peliculas_string = peliculas.join();
  
### Método .split()

- Descripción: Convierte una cadena en un array, utilizando un separador.
- Uso:

        var cadena = "texto1, texto2, texto3";
        var cadena_array = cadena.split(", ");

  
## Ordenar y Revertir un Array

### Método .sort()

- Descripción: Ordena los elementos de un array en orden alfabético (o en orden ascendente si son números).
- Uso:
        
        peliculas.sort();

  
### Método .reverse()

- Descripción: Invierte el orden de los elementos en un array.
- Uso:
        
        peliculas.reverse();
  
### Método find()

- Descripción: Devuelve el primer elemento que cumple con la condición especificada en la función.
- Uso:
        
        var busqueda = lenguajes.find(lenguaje => lenguaje == "PHP");


### Método findIndex()

- Descripción: Devuelve el índice del primer elemento que cumple con la condición especificada en la función. Si no se encuentra, devuelve -1.
- Uso:
    
        var busqueda2 = lenguajes.findIndex(lenguaje => lenguaje == "PHP");
  
### Método some()

- Descripción: Comprueba si al menos un elemento del array cumple con la condición especificada en la función. Devuelve true o false.
- Uso:
        
        var precios = [10, 20, 30, 40, 50];
        var busquedaPrecios = precios.some(precio => precio >= 20); // Devuelve true si hay precios >= 20
        var busquedaPrecios2 = precios.some(precio => precio > 200); // Devuelve false
