
# Funciones 

La función Number y las funciones parseInt y parseFloat en JavaScript se usan para convertir valores a números, pero lo hacen de maneras un poco diferentes. Vamos a explicarlo en palabras sencillas.

### La función Number
La función Number convierte casi cualquier cosa (cadenas de texto, booleanos, etc.) a un número. Intenta convertir el valor entero o decimal de manera directa. Si no puede convertirlo, devolverá NaN (que significa "Not a Number").


            Number("123");  // Devuelve 123 (número)
            Number("123.45");  // Devuelve 123.45 (número decimal)
            Number("hola");  // Devuelve NaN porque "hola" no es un número
            Number(true);  // Devuelve 1 (true se convierte en 1)
            Number(false);  // Devuelve 0 (false se convierte en 0)

### parseInt

parseInt también convierte cadenas de texto a números, pero se enfoca en números enteros (sin decimales). Lee la cadena de texto hasta que encuentra algo que no sea un número y lo convierte hasta ese punto. Si la cadena no empieza con un número, devuelve NaN.


            
            Copiar código
            parseInt("123");  // Devuelve 123 (número entero)
            parseInt("123.45");  // Devuelve 123 (ignora lo que sigue del punto decimal)
            parseInt("123abc");  // Devuelve 123 (ignora "abc")
            parseInt("abc123");  // Devuelve NaN porque no empieza con un número

### parseFloat

Es similar a parseInt, pero maneja números con decimales (flotantes). Convierte la cadena de texto a un número decimal, pero al igual que parseInt, detiene la conversión cuando encuentra algo que no es parte del número.

            Copiar código
            parseFloat("123.45");  // Devuelve 123.45 (número decimal)
            parseFloat("123.45abc");  // Devuelve 123.45 (ignora "abc")
            parseFloat("abc123.45");  // Devuelve NaN porque no empieza con un número


            
### Diferencias clave:

        Number convierte de forma directa: Intenta convertir el valor completo a un número. Si no puede, devuelve NaN.
        
        Ejemplo: Number("123abc") devuelve NaN.


### parseInt y parseFloat leen parcialmente: Convierten la parte numérica inicial de una cadena y se detienen cuando encuentran algo que no es un número.
        
        Ejemplo: parseInt("123abc") devuelve 123.
        parseInt solo devuelve enteros, mientras que parseFloat devuelve números con decimales si los encuentra.
        
        Ejemplo: parseInt("123.45") devuelve 123, pero parseFloat("123.45") devuelve 123.45.

        
### Conclusión:

- Usa Number si quieres convertir algo a un número completo de manera directa.
- Usa parseInt si solo necesitas la parte entera de un número.
- Usa parseFloat si necesitas convertir a un número decimal (flotante).
