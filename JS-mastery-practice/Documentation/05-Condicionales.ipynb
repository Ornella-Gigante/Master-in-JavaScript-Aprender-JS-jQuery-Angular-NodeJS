# Condicional IF 

El condicional if en JavaScript es una forma de tomar decisiones en tu código. Te permite ejecutar diferentes partes del código según si una condición es verdadera o falsa.

## ¿Cómo funciona if?
Imagina que tienes que tomar una decisión: "Si está lloviendo, llevo un paraguas; si no, no lo llevo". En código, esto sería un if en acción.

        Sintaxis básica:
        
        if (condición) {
            // Código a ejecutar si la condición es verdadera
        }

- condición: Es una expresión que puede ser verdadera (true) o falsa (false). Por ejemplo, una comparación como edad >= 18.
- Bloque de código: Si la condición es verdadera, el código dentro de las llaves { } se ejecuta.
Ejemplo sencillo:
        
        let edad = 20;
        
        if (edad >= 18) {
            console.log("Eres mayor de edad");
        }

En este ejemplo:

- La condición es edad >= 18. Si esta condición es verdadera (como en este caso, porque edad es 20), el mensaje "Eres mayor de edad" se mostrará en la consola.
if-else:
- A veces, también quieres hacer algo diferente si la condición es falsa. Aquí es donde entra el else:


        let edad = 16;
        
        if (edad >= 18) {
            console.log("Eres mayor de edad");
        } else {
            console.log("Eres menor de edad");
        }

- En este ejemplo:

Si edad es 16, la condición edad >= 18 es falsa, por lo que se ejecuta el código dentro del else, mostrando "Eres menor de edad".
if-else if-else:
Si tienes más de dos opciones, puedes usar else if para agregar más condiciones:

let temperatura = 30;
        
        if (temperatura > 30) {
            console.log("Hace mucho calor");
        } else if (temperatura > 20) {
            console.log("El clima es agradable");
        } else {
            console.log("Hace frío");
        }

- En este ejemplo:

Si temperatura es 30, la primera condición es falsa (30 > 30), pero la segunda (30 > 20) es verdadera, por lo que se ejecuta "El clima es agradable".
En resumen:
* if: Se usa para tomar decisiones basadas en una condición.
* else: Se ejecuta cuando la condición del if es falsa.
* else if: Se usa cuando tienes múltiples condiciones a evaluar.
