# Bucle for


- El bucle for se utiliza cuando sabes cuántas veces quieres repetir un bloque de código. Se estructura en tres partes: inicialización, condición y actualización.

- Estructura básica:

       
        for (inicialización; condición; actualización) {
            // Código a repetir
        }
        Ejemplo sencillo:
        
        javascript
        Copiar código
        for (var i = 0; i < 5; i++) {
            console.log(i); // Imprime 0, 1, 2, 3, 4
        }

  
- Inicialización (var i = 0): Aquí defines desde qué valor empieza el bucle.
- Condición (i < 5): El bucle se repite mientras esta condición sea verdadera.
- Actualización (i++): Después de cada repetición, el valor de i aumenta en 1.

 
 
# Bucle while

- El bucle while se utiliza cuando no sabes exactamente cuántas veces necesitas repetir el bloque de código, pero sí sabes la condición bajo la cual debe repetirse. Este bucle sigue ejecutándose mientras la condición sea verdadera.

- Estructura básica:
        
        
        while (condición) {
            // Código a repetir
        }
        Ejemplo sencillo:
        
    
        var i = 0;
        while (i < 5) {
            console.log(i); // Imprime 0, 1, 2, 3, 4
            i++; // No olvides actualizar el valor de i para evitar un bucle infinito
        }

  
- Condición (i < 5): El bucle sigue ejecutándose mientras la condición sea verdadera


# Bucle do...while

El bucle do...while es similar al while, pero con una diferencia clave: **el código se ejecuta al menos una vez**, incluso si la condición es falsa desde el principio. Después de ejecutar el bloque de código una vez, verifica la condición para decidir si debe repetirse.

Estructura básica:

        
        do {
            // Código a repetir
        } while (condición);
        Ejemplo sencillo:
        
        javascript
        Copiar código
        var i = 0;
        do {
            console.log(i); // Imprime 0, 1, 2, 3, 4
            i++;
        } while (i < 5);

        
- Ejecuta primero: El bloque de código se ejecuta una vez antes de verificar la condición.


## Diferencias clave:


- for: Usado cuando sabes cuántas veces quieres repetir algo. Tiene una estructura clara con inicialización, condición y actualización.

- while: Usado cuando no sabes cuántas veces repetir algo, pero sabes una condición que debe cumplirse para seguir repitiendo.

- do...while: Similar al while, pero siempre ejecuta el código al menos una vez, incluso si la condición es falsa desde el principio.

## En resumen:

- Usa for cuando tienes un **número definido** de repeticiones.
- Usa while cuando no sabes cuántas veces se repetirá, pero tienes **una condición**.
- Usa do...while cuando necesitas asegurarte de que el código **se ejecute al menos una vez** antes de verificar la condición.
